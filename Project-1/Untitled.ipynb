# Data Analysis and Visualization

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
cancelled_perc

```

```python
plt.figure(figsize=(5,4))
plt.title('Reservation Status Count')
plt.bar(['Not canceled', 'Canceled'],df['is_canceled'].value_counts() , edgecolor = 'k' , width=0.7)
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel', hue='is_canceled', data=df , palette='Blues')
legend_labels,_= ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status in different Hotels',size=20)
plt.xlabel('Hotel')
plt.ylabel('number of reservation')
plt.legend(['non-canceled','canceled'])
plt.show()
```

```python
resort_hotel= df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
a=resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
plt.pie(a , labels=['not-canceled','cancelled'] , autopct='%0.1f%%')
plt.show()
```

```python
city_hotel= df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
b=city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
plt.pie( b , labels=['not-canceled','cancelled'] , autopct='%0.1f%%' )
plt.show()
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.plot(resort_hotel.index , resort_hotel['adr'] , label='Resort Hotel')
plt.plot(city_hotel.index , city_hotel['adr'] , label='City Hotel')
plt.title('Average daily rate of Resort and City Hotel' , fontsize=30)
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
```

```python
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month', hue='is_canceled' , data=df , palette='Blues')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Reservation status per month" , size=20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not-canceled','canceled'])
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.title("ADR per month", fontsize=30)

sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())

plt.show()
```

```python
canceled_data=df[df['is_canceled']==1]
not_canceled_data=df[df['is_canceled']==0]
```

```python

top_10_countries=canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top countries with reservation cancellation',fontsize=30)
plt.pie(top_10_countries , autopct='%0.1f%%' , labels=top_10_countries.index)
plt.show()
```

```python
df['market_segment'].value_counts()
```

```python
df['market_segment'].value_counts(normalize=True)
```

```python
canceled_data['market_segment'].value_counts(normalize=True)
```

```python
canceled_df_adr=canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_df_adr=not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average daily rate')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'], label='Not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'], label='Canceled')
plt.legend()

plt.show()

```

```python
canceled_df_adr=canceled_data.groupby('reservation_status_date')>2016[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_df_adr=not_canceled_data.groupby('reservation_status_date')>2016[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average daily rate')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'], label='Not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'], label='Canceled')
plt.legend()

plt.show()
```
